```python
####################### Task 4 #######################
#Because there isnt a file with the entries, a manual data frame will be made but with different
#Properties as listen below...


#Making the Training data frame

#Home: 1 - HOME, 0 - AWAY
#Top: 1 - IN, 0 - OUT
#Media: 1 - NBC, 2 - ESPN, 3 - FOX, 4 - ABC
#Result: 1 - WIN, 0 - LOSE



Train_1 = {'home': [1, 1, 0, 0, 1, 0],\
          'top': [0, 1, 0, 0, 0, 0], \
          'media':[1, 1, 2, 3, 1, 4],\
          'results':[1, 0, 1, 1, 1, 1]}



#Making the Testing data frame

Test_1 = {'home': [1, 1, 1, 0, 1, 0],\
         'top': [1, 0, 0, 1, 0, 1], \
         'media':[1, 1, 1, 4, 1, 4]}




#Making a matrix data frame

Train1_Df = pd.DataFrame(data = Train_1)
Test1_Df = pd.DataFrame(data = Test_1)
```

```python
################################### PART 1 ##################################

##### Entropy Plot #####

#Making the Decision tree 
Decision_Tree1 = tree.DecisionTreeClassifier(criterion = "entropy")

CL1 = Decision_Tree1.fit(Train1_Df.iloc[:, :3], 
              Train1_Df['results'])

tree.plot_tree(CL1)

DATA1 = tree.export_graphviz(Decision_Tree1, out_file = None, 
                                feature_names = list(Train1_Df.columns.values)[:3],
                                filled = True, 
                                rounded = True, 
                                special_characters = True)



#Rendering the tree and finalizing

graph1 = graphviz.Source(DATA1)
graph1.render("Decision Tree for Entropy")
pred1 = CL1.predict(Test1_Df)
print("Prediction for Entropy:", pred1)




##### GINI Plot #####

Decision_Tree2 = tree.DecisionTreeClassifier()
CL2 = Decision_Tree2.fit(Train1_Df.iloc[:, :3], Train1_Df['results'])
tree.plot_tree(CL2)
DATA_2 = tree.export_graphviz(Decision_Tree2, 
                              out_file = None, 
                              feature_names = list(Train1_Df.columns.values)[:3],
                              filled = True, 
                              rounded = True, 
                              special_characters = True)


graph2 = graphviz.Source(DATA_2)
graph2.render("Decision Tree for Gini:")
pred2 = CL2.predict(Test1_Df)
print("Prediction for Gini:", pred2)
print('End')

############################### END OF PART 1 ###############################
```
